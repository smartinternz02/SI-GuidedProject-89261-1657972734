```python
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.applications.xception import preprocess_input
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(r"D:\datasets\eye disease\preprocessed dataset\preprocessed dataset\training",
                                                 target_size = (299, 299),
                                                 batch_size = 16,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory(r"D:\datasets\eye disease\preprocessed dataset\preprocessed dataset\testing",
                                                 target_size = (299, 299),
                                                 batch_size = 16,
                                                 class_mode = 'categorical')
```

```python
xception = Xception(input_shape=[299,299]+[3] ,weights='imagenet', include_top=False)
```

```python
for layers in xception.layers:
    layers.trainable=False
```

```python
x=Flatten()(xception.output)
```

```python
prediction=Dense(5,activation='softmax')(x)
```

```python
model=Model(inputs=xception.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
r=model.fit_generator(training_set,validation_data=test_set,epochs=50,steps_per_epoch=len(training_set)//32,validation_steps=len(test_set)//32)
```

```python
model.save("diabetic.h5")
```
